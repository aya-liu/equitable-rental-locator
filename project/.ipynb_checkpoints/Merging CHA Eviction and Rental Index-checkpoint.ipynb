```python
%load_ext autoreload
%autoreload 2

import pickle
import pandas as pd
import geopandas
from shapely.geometry import Point
import matplotlib.pyplot as plt
```

```python
# read scraped CHA data
with open("data/CHA_rental_data.obj", "rb") as f:       
    d = pickle.load(f)
cha = pd.DataFrame.from_dict(data = d, orient = "index")
```

```python
# clean CHA data
cols = ['Address','Monthly Rent','Property Type','Bath','Bed',
        'Voucher Necessary','Availability','Contact','URL','Lat','Long']
cha = cha[cols]
cha.Long = -1 * cha.Long

# correct error
cha.loc["4545145", "Long"] = -87.66593 
cha.loc["4545145", "Lat"] = 41.772175
```

```python
cha.head()
```

```python
cha.dtypes
```

```python
# convert to GeoDataFrame
cha['Coordinates'] = list(zip(cha.Long, cha.Lat))
```

```python
cha['Coordinates'] = cha['Coordinates'].apply(Point)
```

```python
gcha = geopandas.GeoDataFrame(cha, geometry='Coordinates')
```

```python
def find_eviction_nums(evictions_df):
    unique_name_array = evictions_df['Name'].unique()
    evictions_dict = {}
    # for index, row in evictions_merge.iterrows():
    for name in unique_name_array:
        filtered_df =  evictions_df[evictions_merge['Name'] == name]
        filtered_df = filtered_df.drop_duplicates(subset= 'GEOID')
        evictions_dict[name] = filtered_df['evictions'].sum()/filtered_df['renter-occupied-households'].sum()
    return evictions_dict
```
